# *Week 1 Analysis*

## Setting Timeline to analyze Stocks

```python
start = datetime.datetime(2021,1,1)
end = datetime.datetime(2021,4,30)
```

## Extracting certain stocks
Data extracted from yahoo

```python
tesla = web.DataReader("TSLA", 'yahoo',start,end)
fastly = web.DataReader("FSLY", 'yahoo',start,end)
zoom = web.DataReader("ZM",'yahoo',start,end)
ibm = web.DataReader("IBM",'yahoo',start,end)
spotify = web.DataReader("SPOT",'yahoo',start,end)
```

## Saving Data gathered into csv files

```python
tesla.to_csv('TeslaStock.csv')
fastly.to_csv('FastlyStock.csv')
zoom.to_csv('ZoomStock.csv')
ibm.to_csv('IBMStock.csv')
spotify.to_csv('SPOTIFYStock.csv')
```

## Gathering Stock Open values from stocks 

```python
tesla['Open'].plot(label = 'Tesla', figsize=(15,7))
fastly['Open'].plot(label = 'Fastly')
zoom['Open'].plot(label = 'Zoom')
ibm['Open'].plot(label = 'IBM')
spotify['Open'].plot(label = 'Spotify')
plt.ylabel('Stock Price')
plt.title('Stock OPEN Prices')
plt.legend()
plt.show()
```

## Data about volume of stocks

```python
tesla['Volume'].plot(label='Tesla',figsize=(15,7))
fastly['Volume'].plot(label='Fastly')
zoom['Volume'].plot(label='Zoom')
ibm['Volume'].plot(label='IBM')
spotify['Volume'].plot(label='Spotify')
plt.legend()
plt.show()
```

# Volatility
## Use of Returns to determine flucuation
By analyzing the returns when can determine the most stable stock out of the group

```python
tesla['returns'] = (tesla['Close']/tesla['Close'].shift(1)) - 1
fastly['returns'] = (fastly['Close']/fastly['Close'].shift(1)) - 1
zoom['returns'] = (zoom['Close']/zoom['Close'].shift(1)) - 1
ibm['returns'] = (ibm['Close']/ibm['Close'].shift(1)) - 1
spotify['returns'] = (spotify['Close']/spotify['Close'].shift(1)) - 1
```

```python
tesla['returns'].hist(bins=50,label='Tesla',alpha=0.5,figsize=(13,6))
fastly['returns'].hist(bins=50,label='Fastly',alpha=0.5)
zoom['returns'].hist(bins=50,label='Zoom',alpha=0.5)
ibm['returns'].hist(bins=50,label='IBM',alpha=0.5)
spotify['returns'].hist(bins=50,label='Spotify',alpha=0.5)
plt.legend()
```

```python
tesla['returns'].plot(kind='kde',label='Tesla',figsize=(15,6))
fastly['returns'].plot(kind='kde',label='Fastly')
zoom['returns'].plot(kind='kde',label='Zoom')
ibm['returns'].plot(kind='kde',label='IBM')
spotify['returns'].plot(kind='kde',label='Spotify')
plt.legend()
```

## *Week 1 Conlusion*

2. Fastly, Zoom, Spotify seemed to be at a dip. Might be good to buy and ride the climb after the fall.
